# About

This dataset contains financial information for the top 500 companies in India, including their market capitalization and quarterly sales. The data is categorized based on market cap and sales quartiles, allowing for detailed analysis and comparison. This dataset can be used to identify trends, patterns, and key metrics that are crucial for understanding the competitive landscape in the Indian market.

# Dataset Features

* Name The name of the company.

* Mar Cap - Crore The market capitalization of the company in crores.
    
* Sales Qtr - Crore The quarterly sales of the company in crores.

* Market_Cap_Category The categorization of the company based on its market cap, with three possible values: "Large Cap", "Mid Cap", "Small


* Sales_Qrt_Category The categorization of the company based on its quarterly sales, with three possible values: "High Sales", "Medium

```python
plt.figure(figsize=(15,8))
Data_Encoded.corr()['Sales Qtr - Crore'].sort_values(ascending = False).head(20).plot(kind='bar',color = 'c')
plt.title('The Top 20 Correlation with Sales Qtr - Crore')
plt.xlabel('Features')
plt.ylabel('Correlation coefficient')

plt.show()
```

```python
fig,ax = plt.subplots(nrows = 1, ncols = 2,figsize=(30,5))
ax[0].hist(Data['Mar Cap - Crore'],log =True, edgecolor = 'black', color = 'b', label = "The Distribution of Mar Cap - Crore")
ax[0].axvline(np.mean(Data['Mar Cap - Crore']), color = 'black', label = "The Mean")
ax[0].axvline(np.median(Data['Mar Cap - Crore']), color = 'black', label = "The Mean")
ax[0].legend()


ax[0].set_title("The Distribution of Mar Cap - Crore")
ax[1].hist(Data['Sales Qtr - Crore'],log=True, edgecolor = 'black', color = 'r', label = "The Distribution of Sales Qtr - Crore")
ax[1].axvline(np.mean(Data['Sales Qtr - Crore']), color = 'black', label = "The Mean")
ax[1].legend()
ax[1].set_title("The Distribution of Sales Qtr - Crore")
plt.show()
                      
              
```

```python
Market_cap_Category = Data["Market_Cap_Category"].value_counts()
Sales_Qrt_Category = Data["Sales_Qrt_Category"].value_counts()
```

```python
fig,ax = plt.subplots(nrows =1, ncols =2, figsize=(20,20))
ax[0].pie(Market_cap_Category.values, labels=Market_cap_Category.keys(), wedgeprops={"edgecolor":"black"},shadow = True, explode = [0,0,0.1,0],autopct='%1.1f%%')
ax[0].set_title("The Distribution of Market_cap")
ax[0].legend()

ax[1].pie(Sales_Qrt_Category.values , labels = Sales_Qrt_Category.keys(), wedgeprops={"edgecolor":"black"} , shadow = True , explode = [0,0,0.1,0] , autopct = '%1.1f%%')
ax[1].set_title("The Distribution of Sales_Qrt")
ax[1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows =1 , ncols = 2, figsize =(15,5))
sns.barplot(ax = ax[0], x=Data["Market_Cap_Category"],y=Data['Mar Cap - Crore'], hue=Data["Market_Cap_Category"],estimator='median',dodge = True)
ax[0].set_title("The Market Capitation")

sns.barplot(ax = ax[1], x=Data["Sales_Qrt_Category"],y=Data['Sales Qtr - Crore'], hue=Data["Sales_Qrt_Category"],estimator='median',dodge = True)
ax[0].set_title("The Sales Quarterly")
plt.show()
```

```python
fig,ax = plt.subplots(nrows = 1,ncols =2 , figsize=(15,5))
sns.barplot(ax = ax[0], x=Data["Market_Cap_Category"],y=Data['Mar Cap - Crore'], hue=Data['Sales_Qrt_Category'],estimator='median',dodge= True)
ax[0].set_title("The Market Capitation VS Sales Quarterly")

sns.barplot(ax = ax[1], x=Data["Sales_Qrt_Category"],y=Data['Sales Qtr - Crore'], hue=Data['Sales_Qrt_Category'],estimator='median',dodge= True)
ax[1].set_title("The Sales Quarterly VS The Market Capitation")
plt.show()
```

```python
sns.jointplot(x=Data['Mar Cap - Crore'], y=Data['Sales Qtr - Crore'], kind='reg', hue_norm = True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(hue=Data['Sales_Qrt_Category'],x=Data['Market_Cap_Category'])
plt.title("The Campany Capacity and them Quarterly Sales")
plt.show()
```

```python
fig, ax = plt.subplots(nrows =2 ,ncols=1, figsize=(20,20))
sns.boxplot(ax= ax[0], y=Data["Mar Cap - Crore"], x=Data["Market_Cap_Category"], hue=Data["Market_Cap_Category"],palette="plasma")
ax[0].set_title("The Boxplot for Market Capacity")

sns.boxplot(ax= ax[1], y=Data["Sales Qtr - Crore"], x=Data["Sales_Qrt_Category"], hue=Data["Sales_Qrt_Category"],palette="plasma")
ax[0].set_title("The Boxplot for Quarterly Sales")
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1, figsize=(20,20))

sns.violinplot(ax= ax[0], y=Data["Mar Cap - Crore"], x=Data["Market_Cap_Category"], hue=Data["Market_Cap_Category"], palette ="plasma")
ax[0].set_title("The Violinplot for Market Capacity")

sns.violinplot(ax= ax[1], y=Data["Sales Qtr - Crore"], x=Data["Sales_Qrt_Category"], hue=Data["Sales_Qrt_Category"],palette ="plasma")
ax[1].set_title("The Violinplot for Quarterly Sales")
plt.show()
```
